## Data manipulation

### Website address length

```python
df_www = with_website[['Nazwa placówki', 'Adres www']].copy()
```

```python
# Website address length
df_www['Długość adresu www'] = df_www['Adres www'].map(lambda x: len(x))
df_www
```

```python
df_www_length = df_www.copy()
df_www_length.drop('Adres www', axis=1, inplace=True)
df_www_length
```

## Grouping by school type

```python
df['Typ podmiotu'].unique()
```

```python
df_counts = df.groupby('Typ podmiotu').count().sort_values('Rok szkolny', ascending=False).head(8)['Rok szkolny']
df_counts = df_counts.rename('Liczba placówek')

others_count = df.shape[0] - df_counts.sum()
# others_count
# df_counts.info()
others_series = pd.Series([others_count], index=['Inne'], name='Liczba placówek')
# others_new
df_counts = df_counts.add(others_series, fill_value=0).sort_values(ascending=False)

df_counts
```

```python
plt.pie(df_counts, labels=df_counts.index, autopct='%1.0f%%')
```
